```python
import requests
```

```python
url = 'https://www.flipkart.com/samsung-galaxy-z-flip4-5g-bora-purple-128-gb/p/itm0e383b143be79?pid=MOBGHFGDHQWKX6HK&lid=LSTMOBGHFGDHQWKX6HKSQA6BM&marketplace=FLIPKART&q=samsung+galaxy+z+flip+4+purple&store=tyy%2F4io&srno=s_1_1&otracker=AS_Query_OrganicAutoSuggest_1_24_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_24_na_na_na&fm=Search&iid=ecff3192-2934-49c1-b3c8-eaeffe05d73d.MOBGHFGDHQWKX6HK.SEARCH&ppt=sp&ppn=sp&ssid=jj6pt43eq80000001675091982220&qH=6125aa1e45b5c491'
```

```python
r = requests.get(url)
```

```python
htmlcontent = r.content
```

```python
htmlcontent
```

```python
#Beautiful Soup converts unstructured data into structured and cleaned, understandable format. 
soup = bs(htmlcontent,'html.parser')
print(soup.prettify())
```

### Display Title of Product

```python
title = soup.title
print(title.string)
```

### Check price of the product

```python
price = soup.find_all('div', class_ = '_30jeq3 _16Jk6d')
product_price = []

for i in range(0,len(price)):
    product_price.append(price[i].get_text())
```

```python
product_price
```

### Display customer names

```python
names = soup.find_all('p', class_='_2sc7ZR _2V5EHH')
customer_names = []
for i in range(0,len(names)):
    customer_names.append(names[i].get_text())
customer_names
```

### Customer comments

```python
comments = soup.find_all('p',class_='_2-N8zT')
customer_comments = []
for i in range(0,len(comments)):
    customer_comments.append(comments[i].get_text())
customer_comments
```

### Stars given by customers

```python
stars = soup.find_all('div', class_ = '_3LWZlK _1BLPMq')
customer_stars = []
for i in range(0,len(stars)):
    customer_stars.append(stars[i].get_text())
customer_stars
```

### Customer reviews

```python
reviews = soup.find_all('div', class_ = 't-ZTKy')
customer_reviews = []
for i in range(0,len(reviews)):
    customer_reviews.append(reviews[i].get_text())
customer_reviews
```

### Convert data into tabular format 

```python
customer_stars.append("1")
```

```python
import pandas as pd
```

```python
len(customer_stars)
```

```python
df = pd.DataFrame()
df['Customer Names'] =customer_names
df['Customer Stars'] =customer_stars
df['Customer Comments']=customer_comments
df['Customer Reviews'] =customer_reviews
df
```

```python
df.to_csv('SamsungFlip4.csv')
```

Q. Explain the concept of web scrapping also explain beautitufsoup functioning by extracting /scraping data from a url.

Web scraping is the process of extracting content and data from a website.

Scraping a web page means requesting specific data from a target webpage. When we scrape a page, the code we write sends our request to the server hosting the destination page. The code then downloads the page only extracting the elements of the page defined initially in the crawling job.

Beautiful Soup :- 

Beautiful Soup is a python library in Web Scrapping used to display web content in cleaned format. Beautiful Soup provides simple methods for navigating, searching, and modifying a parse tree in HTML, XML files. It transforms a complex HTML document into a tree of Python objects. It also automatically converts the document to Unicode
